```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq
```

```python
# Carica i dati dal file CSV
data = np.loadtxt('testFft2.csv', delimiter=',', skiprows=1)  # Sostituisci con il tuo file CSV
sampling_rate = 8e6
sonda1 = data[:, 0]
sonda2 = data[:, 1]
```

```python
# Calcola la FFT
N = len(sonda2)
fft_values = fft(sonda2)
fft_freqs = fftfreq(N, 1/sampling_rate)
```

```python
# Ignora la componente DC (frequenza 0)
fft_values_no_dc = fft_values[1:]  # Escludi il primo valore della FFT
fft_freq_no_dc = fft_freqs[1:]  # Escludi la frequenza 0
# Limita le frequenze tra 0 e 1 kHz
mask = (fft_freq_no_dc >= 0) & (fft_freq_no_dc <= 1000)
```

```python
# Plot del diagramma spettrale
plt.plot(fft_freqs_no_dc[mask], np.abs(fft_values_no_dc[mask]))
plt.xlabel('Frequenza (Hz)')
plt.ylabel('Ampiezza')
plt.grid(True)
plt.show()
```
